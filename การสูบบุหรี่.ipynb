## Preprocessing

```python
# Rename the column 'ภาค' to 'region'
df.rename(columns={'ภาค': 'region'}, inplace=True)


# Replace values in the "ภาค" column
replacement_dict = {
    'ภาคเหนือ': 'North',
    'ภาคตะวันออกเฉียงเหนือ': 'Eastern',
    'ภาคกลาง': 'Central',
    'ภาคใต้': 'South'
}
df['region'] = df['region'].replace(replacement_dict)

df
```

```python
df['พฤติกรรมการสูบบุหรี่'].value_counts()
```

```python
# Rename the column 'ภาค' to 'region'
df.rename(columns={'พฤติกรรมการสูบบุหรี่': 'bahavior'}, inplace=True)

# Replace values in the "ภาค" column
replacement_dict = {
    'ผู้ที่มีพฤติกรรมสูบบุหรี่เป็นประจำ': 'Frequent',
    'ผู้สูบบุหรี่อายุต่ำกว่า 15 ปี': 'Under 15',
    'ผู้สูบบุหรี่ในวัยรุ่นอายุ 15-18 ปี': '15-18',
    'ผู้ที่อายุ 60 ปีขึ้นไป ที่สูบบุหรี่': '60+ smoke',
    'อยู่ในสถานที่ที่มีผู้สูบบุหรี่ หรือมีโอกาสได้รับควันบุหรี่เป็นประจำในทุกวัน': 'in situation',
    'วัยทำงานอายุ 15-59 ปี ที่สูบบุหรี่เป็นประจำ': '15 Frequent'
}
df['bahavior'] = df['bahavior'].replace(replacement_dict)
df
```

```python
# สร้างกราฟแสดงแนวโน้ม
plt.figure(figsize=(10, 6))
for region, group in df.groupby('region'):
    plt.plot(group['bahavior'], group['จำนวนผู้สูบบุหรี่'], label=region)

plt.xlabel('bahavior')
plt.ylabel('Total')
plt.title('Trend by bahavior and region')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Calculate the total number of smokers across all regions
total_smokers = df['จำนวนผู้สูบบุหรี่'].sum()

# Grouping by "ภาค" and calculating the percentage of smokers within each group
percentage_smokers = df.groupby("region")["จำนวนผู้สูบบุหรี่"].sum() / total_smokers * 100

print(percentage_smokers)

# Calculate total number of smokers for each region
region_totals = df.groupby('region')['จำนวนผู้สูบบุหรี่'].sum()

plt.figure(figsize=(6, 6))
plt.pie(region_totals, labels=region_totals.index, autopct='%1.1f%%', startangle=140)

# Draw a circle in the center of pie to make it look like a donut
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title('Total')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()
plt.show()
```

```python
# Grouping by "ภาค" and summing the values of "จำนวนคนสูบบุหรี่" within each group
grouped_data = df.groupby("region")["จำนวนผู้สูบบุหรี่"].sum()

print(grouped_data)

# Create bar plot
plt.figure(figsize=(10, 6))
plt.bar(grouped_data.index, grouped_data.values, color='skyblue')
plt.xlabel('region')
plt.ylabel('Total')
plt.title('Totalbyregion')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Rename the column 'เพศ' to 'region'
df.rename(columns={'เพศ': 'Gender'}, inplace=True)

# Replace values in the "ภาค" column
replacement_dict = {
    'ชาย': 'Male',
    'หญิง': 'Female'
}
df['Gender'] = df['Gender'].replace(replacement_dict)
df
```

```python
# Grouping by "region" and "เพศ" and summing the values of "จำนวนผู้สูบบุหรี่" within each group
grouped_data = df.groupby(["Gender"])["จำนวนผู้สูบบุหรี่"].sum()

print(grouped_data)

# Create bar plot
plt.figure(figsize=(10, 6))
plt.bar(grouped_data.index, grouped_data.values, color='skyblue')
plt.xlabel('Gender')
plt.ylabel('Total')
plt.title('Total by Gender')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Grouping by "region" and "เพศ" and summing the values of "จำนวนผู้สูบบุหรี่" within each group
grouped_data = df.groupby(["region", "Gender"])["จำนวนผู้สูบบุหรี่"].sum().unstack()

# Create bar plot
plt.figure(figsize=(10, 6))
grouped_data.plot(kind='bar', color=['skyblue', 'orange'])
plt.xlabel('Region')
plt.ylabel('Total')
plt.title('Number of smokers by region and gender')
plt.xticks(rotation=45)
plt.tight_layout()
plt.legend(title='Gender', loc='upper right')
plt.show()
```
