```python
# สร้าง DataFrame
import pandas as pd
data = {'พฤติกรรมการสูบบุหรี่': ["วัยทำงานอายุ 15-59 ปี ที่สูบบุหรี่เป็นประจำ", "วัยทำงานอายุ 15-59 ปี ที่สูบบุหรี่เป็นประจำ"]}
df = pd.DataFrame(data)

# สร้างฟังก์ชันที่ใช้ .replace() บนแต่ละสตริงของลิสต์
def replace_string(s):
    return s.replace("วัยทำงานอายุ 15-59 ปี ที่สูบบุหรี่เป็นประจำ", "อายุ 15-59 ปีสูบประจำ","ผู้ที่มีพฤติกรรมสูบบุหรี่เป็นประจำ","สูบเป็นประจำ")

# ใช้ apply เพื่อใช้ฟังก์ชันกับทุกสตริงในคอลัมน์
df['พฤติกรรมการสูบบุหรี่'] = df['พฤติกรรมการสูบบุหรี่'].apply(replace_string)

# แสดง DataFrame หลังจากการเปลี่ยนข้อมูล
print(df)
```
